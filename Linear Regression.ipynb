# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
%matplotlib inline
```

```python
cmp = pd.read_csv("C:\\Users\\KIRTI PRATIHAR\\JUPYTER IDIOT\\Machine Learning Algorithm\\1000_Companies.csv")
x= cmp.iloc[:,:-1].values
y=cmp.iloc[:,4].values
cmp.head()
```

## Data Visualisation
## Building the correlation Matrix

```python

numeric_data = cmp.select_dtypes(include=['float64', 'int64'])
#sns.heatmap(cmp.corr())-this will give error as there are non numeric data present
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=True)
corr_matrix = numeric_data.corr()
#sns.heatmap(corr_matrix, annot=True, cmap='coolwarm') cmap- chooses the tone
```

## Encoding categorical data

```python

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
labelencoder = LabelEncoder()
x[:,3]=labelencoder.fit_transform(x[:,3])#this will covert state like new york into numbers

# Use OneHotEncoder with ColumnTransformer for one-hot encoding
ct = ColumnTransformer(
    transformers=[('onehot', OneHotEncoder(), [3])],  # Apply OneHotEncoder to column index 3
    remainder='passthrough'  # Keep other columns unchanged
)

# Transform and convert to NumPy array
x = np.array(ct.fit_transform(x))
print(x)
```

## Avoid Dummy Variable

```python
x=x[:,1:]
```

## Splitting the data into training set and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

## Fitting Multiple Linear Regression to the Training set

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

## Predicting the test set results

```python
y_pred =reg.predict(x_test)
y_pred
```

## Calculate the cofficients 

```python
print(reg.coef_)
```

## Calculate the intersept

```python

print(reg.intercept_)
```

## calculate the R squared value

```python

from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
